```python
#(a)

def jacobi_iteration_matrix(A):
    D = np.diag(np.diag(A))
    L = -np.tril(A, k=-1)
    U = -np.triu(A, k=1)
    inv_D = np.linalg.inv(D)
    TJ = np.dot(inv_D, L + U)
    return TJ

def gauss_seidel_iteration_matrix(A):
    D = np.diag(np.diag(A))
    L = -np.tril(A, k=-1)
    U = -np.triu(A, k=1)
    D_L_inv = np.linalg.inv(D + L)
    TGS = np.dot(D_L_inv, U)
    return TGS

# Given matrix A
A = np.array([[4, 1], [1, 4]])

b = [5,5]


# Calculate Jacobi and Gauss-Seidel iteration matrices
TJ = jacobi_iteration_matrix(A)
TGS = gauss_seidel_iteration_matrix(A)

# Print the results
print("Jacobi Iteration Matrix (TJ):")
print(TJ)

print("\nGauss-Seidel Iteration Matrix (TGS):")
print(TGS)

#(b)

#Find infinity norm of TJ and TGS
print("Infinity norm of TJ:")
print(np.linalg.norm(TJ,ord=np.inf))
print("Infinity norm of TGS:")
print(np.linalg.norm(TGS,ord=np.inf))

#Find the spectral radius of TJ and TGS
print("Spectral radius of TJ:")
print(max(abs(np.linalg.eigvals(TJ))))
print("Spectral radius of TGS:")
print(max(abs(np.linalg.eigvals(TGS))))
```

```python
#(c)

#Jacobi method
import numpy as np

def J(A, b, x):
    A = np.array(A)
    n = len(A)
    InvD = np.zeros([n, n])
    R = A
    for i in range(n):
        InvD[i, i] = 1 / A[i, i]
        R[i, i] = 0
    return np.matmul(InvD, (b - np.matmul(R, x)))

A = [[4, 1], [1, 4]]
b = [5, 5]
x = [[0, 0]]
MaxIter = 5
ErrorTable = [1]
ErrorRatio = [0]
eps = 0.001
i = 1

while i <= MaxIter and abs(ErrorTable[i - 1]) > eps:
    xi = J(A, b, x[i - 1])
    x.append(xi)
    ei = np.linalg.norm(x[i] - x[i - 1], np.inf)  # /np.linalg.norm(x[i],np.inf)
    ErrorTable.append(ei)
    zi = np.linalg.norm(x[i] - x[i - 1], np.inf) / np.linalg.norm(x[i - 1] - x[i - 2], np.inf)
    ErrorRatio.append(zi)
    i += 1

print("Solution: Jacobi method")
'''
column 1: k (iteration step)
column 2: (1st component of computed solution vector at step k)
column 3: (2nd component of computed solution vector at step k)
column 4: ke(k)k∞ (error norm at step k)
column 5: ke(k)k∞/ke(k−1)k∞ (ratio of successive error norms at step k)

'''
from tabulate import tabulate

table = []
for i in range(len(x)):
    table.append([i, x[i][0], x[i][1], ErrorTable[i], ErrorRatio[i]])

print(tabulate(table, headers=["k", "x1", "x2", "ke(k)k∞", "ke(k)k∞/ke(k−1)k∞"]))
```

```python
#Gauss Siedel Method
from numpy import linalg
def GS(A, b, x):
  A = np.array(A)
  n = len(A)
  xnew = np.zeros(n)
  for i in range(n):
    xnew[i] = (b[i] - sum([A[i, j]*x[j] for j in range(i+1, n)]) - sum([A[i, j]*xnew[j] for j in range(i)]))/A[i, i]
  return xnew

A = [[4, 1], [1, 4]]
b = [5,5]
x = [[0, 0]]
MaxIter = 100
ErrorTable = [1]
ErrorRatio =[0]
eps = 0.001
i = 1

while i <= MaxIter and abs(ErrorTable[i - 1]) > eps:
  xi = GS(A, b, x[i - 1])
  x.append(xi)
 # ei = np.linalg.norm(x[i] - x[i - 1])/np.linalg.norm(x[i])
  ei=np.linalg.norm(x[i]-x[i-1],np.inf)#/np.linalg.norm(x[i],np.inf)
  ErrorTable.append(ei)
  zi = np.linalg.norm(x[i] - x[i - 1], np.inf) / np.linalg.norm(x[i - 1] - x[i - 2], np.inf)
  ErrorRatio.append(zi)
  i+=1
# Gause Seidel Solution
print("Solution: Gauss-Seidel method")
'''
column 1: k (iteration step)
column 2: (1st component of computed solution vector at step k)
column 3: (2nd component of computed solution vector at step k)
column 4: ke(k)k∞ (error norm at step k)
column 5: ke(k)k∞/ke(k−1)k∞ (ratio of successive error norms at step k)

'''

from tabulate import tabulate

table = []
for i in range(len(x)):
    table.append([i, x[i][0], x[i][1], ErrorTable[i], ErrorRatio[i]])

print(tabulate(table, headers=["k", "x1", "x2", "ke(k)k∞", "ke(k)k∞/ke(k−1)k∞"]))
```

#### **Which method converges faster? Why?**
##### Gauss-Seidel converges faster than Jacobi, because Gauss-Seidel uses the most recent values of the iterates to compute the next iterate, while Jacobi uses the previous values of the iterates to compute the next iterate.
